#**pytorch model**

####__init__ method is the constructor of the LinearRegression class.super().__init__() calls the constructor of the parent class (nn.Module), initializing it properly.self.weights and self.bias are parameters of the linear regression model nn.Parameter: Wraps a tensor as a parameter that is tracked by the nn.Module.torch.randn(1, requires_grad=True, dtype=torch.float): Initializes a tensor of shape (1,) with random values drawn from a standard normal distribution. requires_grad=True indicates that gradients will be computed for this tensor during backpropagation.##

````python
import torch
from torch import nn

class LinearRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1, requires_grad=True, dtype=torch.float))
        self.bias = nn.Parameter(torch.randn(1, requires_grad=True, dtype=torch.float))

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.weights * x + self.bias
````

forward method defines the forward pass of the model.
x: torch.Tensor specifies that x should be a tensor of type torch.Tensor.
-> torch.Tensor specifies that the method returns a tensor of type torch.Tensor.
The method computes the linear transformation self.weights * x + self.bias, which is the core computation of a linear regression model.
self.weights * x: Element-wise multiplication of x (input data) with self.weights.
+ self.bias: Addition of self.bias (bias term).
The result (self.weights * x + self.bias) is the predicted output of the linear regression model given input x.

````python
loss_fn = nn.L1Loss()
optimizer = torch.optim.SGD(params = model.parameters(),lr = 0.01) # adjusts parameters to reduce loss
````

Screenshot (1).png

**In PyTorch, gradients are accumulated by default. When you call .backward(), the computed gradients are added to the .grad attribute of each parameter. This behavior is intentional, as it allows for gradient accumulation over multiple backward passes (useful in certain scenarios like batch accumulation).

If you don't clear the gradients using optimizer.zero_grad():

The gradients from previous steps will interfere with the current computation.
This leads to incorrect parameter updates and poor model performance.**

````python
epochs = 1000 # hyperparameters

for epoch in range(epochs) :  # training loop
  model.train()
  #forward pass
  y_pred = model(x_train)
  loss = loss_fn(y_pred,y_train)
  optimizer.zero_grad()   # zero gradients
  loss.backward()  #backpropagation
  optimizer.step() # perform gradient descent
print(f"loss is {loss} ")
print(f"EPOCH {epochs} || TRAINING LOSS {loss}")
model.eval()   # testing
````

````python
import torch
from torch import nn

# Constructing nn module
class Circle(nn.Module):
    def __init__(self):
        super().__init__()
        # Constructing fully connected layers
        self.layer_1 = nn.Linear(in_features=3, out_features=5)
        self.layer_2 = nn.Linear(in_features=5, out_features=1)

    # Forward method
    def forward(self, x):
        return self.layer_2(self.layer_1(x))

# Define the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Instantiate and move model to device
model2 = Circle().to(device)
print(model2)
````
